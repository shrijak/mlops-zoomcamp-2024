### Importing libraries

```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### Q1. Downloading the data

```python
df = pd.read_parquet('yellow_tripdata_2023-01.parquet')
df
```

### Q1. Answer
There are 19 columns in the dataframe

### Q2. Computing duration

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
df
```

```python
df.duration.describe()
```

### Q2. Answer 
The standard deviation of the trips duration in January is 42.594 minutes

### Q3. Dropping outliers

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
df
```

### Q3. Answer
No. of rows before: 3066766, after: 3009173. So, fraction of the records left after outliers are dropped is 98.12%

### Q4. One-hot encoding

```python
categorical = ['PULocationID', 'DOLocationID']
df[categorical] = df[categorical].astype(str)
train_dicts = df[categorical].to_dict(orient='records')
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
len(dv.feature_names_)
```

### Q4. Answer
The dimensionality of the feature matrix is 515

### Q5. Training a model

```python
target = 'duration'
y_train = df[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)

mean_squared_error(y_train, y_pred, squared=False)
```

### Q5. Answer
The RMSE of the model on the training data is 7.649

### Q6. Evaluating the model

```python
df_val = pd.read_parquet('yellow_tripdata_2023-02.parquet')
df_val['duration'] = df_val.tpep_dropoff_datetime - df_val.tpep_pickup_datetime
df_val.duration = df_val.duration.apply(lambda td: td.total_seconds() / 60)
df_val = df_val[(df_val.duration >= 1) & (df_val.duration <= 60)]
df_val[categorical] = df_val[categorical].astype(str)
val_dicts = df_val[categorical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
y_val = df_val[target].values
y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```

### Q6. Answer
The RMSE of the model on the validation data is 7.811
